```python
import json 
import os 
```

```python
file_path = './../outputs/time-parse/'
files = os.listdir('./../outputs/time-parse')
```

```python
file = files[5]
data = json.load(open(file_path + file))
```

```python
dims = set()
for i in data:
    d = data[i]['time-raw']
    for j in d:
        dims.add(j['dim'])
```

```python
kk = []
for i in data:
    d = data[i]['time-raw']
    for j in d:
        if j['dim'] == 'time' or j['dim'] == 'duration':
            kk.append(data[i])
```
